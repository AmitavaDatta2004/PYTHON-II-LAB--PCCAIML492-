# EDA Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
data=pd.read_csv("iris.csv")
```

```python
print(data.head())
```

```python
print(data)
```

```python
print(data.describe())
```

```python
print(data.info())
```

```python
plt.plot(data['variety'])
plt.xlabel('no of data points')
plt.ylabel('variety')
plt.show()
```

```python
plt.plot(data['sepal.length'])
plt.show()
```

```python
plt.hist(data['variety'])
plt.show()
```

```python
plt.hist(data['sepal.width'])
plt.show()
```

```python
sb.scatterplot(data,x='sepal.length',y='sepal.width',hue='variety')
# plt.plot()
plt.show()
```

```python
sb.pairplot(data,hue='variety')
plt.show()
```

```python
sb.FacetGrid(data,hue='variety').map(sb.histplot,'petal.length',kde=True).add_legend()
```
