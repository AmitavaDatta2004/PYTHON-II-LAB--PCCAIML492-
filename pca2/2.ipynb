```python
plt.hist(data['variety'])
plt.show()
plt.hist(data['sepal.length'])
plt.show()
plt.hist(data['sepal.width'])
plt.show()
plt.hist(data['petal.width'])
plt.show()
```

ScatterPlot

```python
sb.scatterplot(data=data,x="sepal.width",y="sepal.length",hue="variety")
plt.show()
```

PairPlot

```python
data=pd.read_csv('iris.csv')
sb.pairplot(data=data,hue="variety")
plt.show()
```

Histogram and Density Curve

```python
sb.FacetGrid(data=data,hue="variety").map(sb.histplot,'petal.length',kde=True).add_legend()
```

```python
sb.FacetGrid(data=data,hue="variety").map(sb.histplot,"petal.length",kde=True).add_legend()
plt.show()
```

Box Plot

```python
sb.boxplot(data=data,x="variety",y="sepal.length")
plt.grid()
plt.show()
```

Violin Plot

```python
sb.violinplot(data=data,x="variety",y="sepal.length")
plt.show()
```
