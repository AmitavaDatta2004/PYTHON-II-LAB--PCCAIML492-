```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load the iris dataset from the CSV file
df = pd.read_csv('iris.csv')  # Replace with the actual file path

# Extract features (sepal and petal dimensions)
data = df[['sepal.length', 'sepal.width', 'petal.length', 'petal.width']].values

# Apply K-Means clustering (3 clusters for 3 species)
kmeans_model = KMeans(n_clusters=3).fit(data)
labels = kmeans_model.labels_
clusters = kmeans_model.cluster_centers_

# Plotting the results for the first two features (sepal length and width)
plt.scatter(data[:, 0], data[:, 1], c=labels, label='Data Points')
plt.scatter(clusters[:, 0], clusters[:, 1],marker="X",color='red',s=200, label='Centers')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend()
plt.grid()
plt.show()
```
