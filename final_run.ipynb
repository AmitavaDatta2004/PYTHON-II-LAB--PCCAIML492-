```python
health=pd.read_csv('health.csv')
```

```python
print(health)
```

```python
health.drop(columns=['ID'],inplace=True)
```

```python
health=health.to_numpy()
```

```python
print(health)
```

```python
list=[]
for i in range(1,11):
    model=KMeans(n_clusters=i,random_state=1).fit(health)
    labels=model.labels_
    clusters=model.cluster_centers_
    inertia=model.inertia_
    list.append(inertia)
print(list)
```

```python
plt.plot(list)
plt.grid()
plt.show()
```

```python
model=KMeans(n_clusters=3,random_state=1).fit(health)
labels=model.labels_
clusters=model.cluster_centers_
plt.scatter(health[:,1],health[:,2],c=labels,label='No of data point')
plt.scatter(clusters[:,1],clusters[:,2],color='red',marker='X',label='Clusters')
plt.legend()
plt.show()
```

```python
iris=load_iris()
flower=iris.data
print(flower)
```

```python
l=[]
for i in range(1,11):
    model=KMeans(n_clusters=i,random_state=1).fit(flower)
    clusters=model.cluster_centers_
    labels=model.labels_
    inertia=model.inertia_
    l.append(inertia)
    plt.scatter(flower[:,0],flower[:,1],c=labels,label="Data")
    plt.scatter(clusters[:,0],clusters[:,1],color='red',marker='X',label='cluster')
    plt.legend()
    plt.show()

plt.plot(l)
plt.grid()
plt.show()
```
